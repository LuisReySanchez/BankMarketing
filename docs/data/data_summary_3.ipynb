## Variables individuales

En esta secci√≥n se presenta un an√°lisis detallado de cada variable individual. Se muestran estad√≠sticas descriptivas, gr√°ficos de distribuci√≥n y de relaci√≥n con la variable objetivo (si aplica). Adem√°s, se describen posibles transformaciones que se pueden aplicar a la variable

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 8)
sns.countplot(x = 'age', data = df)
ax.set_xlabel('Age', fontsize=15)
ax.set_ylabel('Count', fontsize=15)
ax.set_title('Age Count Distribution', fontsize=15)
sns.despine()
```

```python
indep_variables = df.select_dtypes('object')
indep_variables = [x for x in indep_variables.columns if x != 'y']
```

```python
for col in indep_variables:
    counts = df[col].value_counts().sort_index()
    if len(counts) > 10:
      fig = plt.figure(figsize=(30, 10))
    else:
      fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    counts.plot.bar(ax = ax, color='steelblue')
    ax.set_title(col + ' counts')
    ax.set_xlabel(col) 
    ax.set_ylabel("Frequency")
plt.show()
```

Las variables categoricas ser√°n transformadas utilizando alguna representaci√≥n numerica. Puede ser one hot encoding o label encondig.

Adicionalmente, se transformar√° la edad para dejarla como un rango, es decir. si la persona tiene 46 a√±os solo se tomar√° el primer digito siendo 4; si la persona tiene 29 a√±os se tomara el 2.
